```python
louisville_crime_2020.rename(columns={'INCIDENT_NUMBER': 'Incident_Number', 'DATE_REPORTED': 'Date_Reported', 'DATE_OCCURED': 'Date_Occurred', 'CRIME_TYPE': 'Crime_Type', 'UOR_DESC': 'UOR_Offense_Description', 'NIBRS_CODE': 'NIBRS_Code', 'UCR_HIERARCHY': 'UCR_Hierarchy', 'ATT_COMP': 'Att_Comp', 'LMPD_DIVISION': 'LMPD_Division', 'LMPD_BEAT': 'LMPD_Beat', 'PREMISE_TYPE': 'Premise_Type', 'BLOCK_ADDRESS': 'Block_Address', 'ZIP_CODE': 'ZIP_Code'}, inplace=True)
louisville_crime_2020.columns
```

```python
louisville_crime_2020['Date_Reported'] = pd.to_datetime(
    louisville_crime_2020['Date_Reported'],
    errors='coerce'
)

louisville_crime_2020['Date_Occurred'] = pd.to_datetime(
    louisville_crime_2020['Date_Occurred'],
    errors='coerce'
)
louisville_crime_2020.dtypes
```

```python
louisville_crime_2020['occurred_year'] = louisville_crime_2020['Date_Occurred'].dt.year
louisville_crime_2020['occurred_month'] = louisville_crime_2020['Date_Occurred'].dt.month
louisville_crime_2020['occurred_day'] = louisville_crime_2020['Date_Occurred'].dt.day
louisville_crime_2020['occurred_hour'] = louisville_crime_2020['Date_Occurred'].dt.hour
louisville_crime_2020['occurred_minute'] = louisville_crime_2020['Date_Occurred'].dt.minute
louisville_crime_2020['occurred_dow'] = louisville_crime_2020['Date_Occurred'].dt.day_name()
louisville_crime_2020.head()
```

```python
cols_to_move = ['Date_Occurred', 'occurred_year', 'occurred_month', 'occurred_day', 'occurred_hour', 'occurred_minute', 'occurred_dow']

cols = list(louisville_crime_2020.columns)

for col in cols_to_move:
    cols.remove(col)

insert_at = 8
for i, col in enumerate(cols_to_move):
    cols.insert(insert_at + i, col)

louisville_crime_2020 = louisville_crime_2020[cols]
louisville_crime_2020.head()
```

```python
louisville_crime_2020['ZIP_Code'] = (
    louisville_crime_2020['ZIP_Code']
    .astype(str)
    .str.replace(r'\.0$', '', regex=True)
)
louisville_crime_2020.head()
```

```python
louisville_crime_2020['ZIP_Code'] = (
    louisville_crime_2020['ZIP_Code']
    .replace('40056', '40201')
)

louisville_crime_2020
```

```python
louisville_crime_2020['City'] = (
    louisville_crime_2020['City']
    .replace('LOUUSVILLE', 'LOUISVILLE')
)
louisville_crime_2020
```

```python
louisville_crime_2020['Att_Comp'] = (
    louisville_crime_2020['Att_Comp']
    .replace('', pd.NA)
    .fillna('UNKNOWN')
)
louisville_crime_2020
```
